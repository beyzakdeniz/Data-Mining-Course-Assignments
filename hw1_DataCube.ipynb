### Task 5: Compute a Data Cube 

Computing a data cube from the dataset. You can use pandas to create a multi-dimensional array and perform various computations.

```python
# Merge all DataFrames to form a single DataFrame
merged_df = pd.merge(product_hierarchy_df, sales_df, on='product_id')
merged_df = pd.merge(merged_df, store_cities_df, on='store_id')



# Convert the date column to datetime
merged_df['date'] = pd.to_datetime(merged_df['date'])

merged_df['year'] = merged_df['date'].dt.year
merged_df['month'] = merged_df['date'].dt.month
merged_df['day'] = merged_df['date'].dt.day

# # Drop rows where 'sales' or 'stock' have NaN values
# cleaned_df = merged_df.dropna(subset=['sales', 'stock'])

merged_df = merged_df.dropna(subset=['sales'])
merged_df = merged_df[merged_df['sales'] != 0.0]


# Creating the data cube
data_cube = pd.pivot_table(
    merged_df,
    values=['stock', 'sales'],
    index=['year', 'month', 'day'],
    columns=['category', 'sub_category', 'store_name'],
    fill_value=0,
    aggfunc='sum'
)


# Display the data cube
data_cube.head()
```

### Task 6: Aggregations and Filtering Operations on Data Cube

1. Display Total sales amount by year and month
2. Display Sales amount for a specific category ('H03') and sub-category('H0314')

```python
# Aggregation: Total sales amount by year and month
agg_sales_by_year_month = merged_df.groupby(['year', 'month'])['sales'].sum().reset_index()

# Filtering: Sales amount for a specific category and sub-category
filtered_sales = merged_df[(merged_df['category'] == 'H03') & (merged_df['sub_category'] == 'H0314')]
agg_filtered_sales = filtered_sales.groupby(['year', 'month'])['sales'].sum().reset_index()

# Display results
print("Aggregation: Total Sales Amount by Year and Month")
display(agg_sales_by_year_month.head())

print("\nFiltering: Sales Amount for H03 > H0314")
display(agg_filtered_sales.head())
```
